1. Bagaimana tren konsentrasi PM2.5 di berbagai stasiun dari tahun 2013 hingga 2017?
2. Bagaimana hubungan antara faktor meteorologi (suhu, tekanan udara, kecepatan angin) dengan tingkat polusi PM2.5?


```python
%pip install -r requirements.txt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import os
import glob
```

```python
folder_path = "air_quality_datasets"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Cek ketersediaan data
if not csv_files:
    raise ValueError("Tidak ditemukan file CSV di folder yang dimaksud")

# Gabungkan semua file
df_list = []
for file in csv_files:
    if os.path.getsize(file) > 0:  # Skip file kosong
        temp_df = pd.read_csv(file)
        df_list.append(temp_df)

combined_df = pd.concat(df_list, ignore_index=True)

# Cek struktur data
print("Informasi Dataset:")
combined_df.info()

print("\n5 Data Pertama:")
combined_df.head()
```

```python
# Handle missing values
print("Missing Values Sebelum Cleaning:")
print(combined_df.isnull().sum())

# Isi missing values PM2.5 dengan median per stasiun
combined_df['PM2.5'] = combined_df.groupby('station')['PM2.5'].transform(
    lambda x: x.fillna(x.median())
)

# Buat kolom datetime
combined_df['date'] = pd.to_datetime(
    combined_df[['year', 'month', 'day', 'hour']].astype(str).agg('-'.join, axis=1) + ':00',
    format='%Y-%m-%d-%H:%M'
)

# Hapus duplikat
combined_df = combined_df.drop_duplicates(subset=['station', 'date'])

print("\nMissing Values Setelah Cleaning:")
print(combined_df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Statistik deskriptif
print("Statistik Deskriptif PM2.5:")
print(combined_df['PM2.5'].describe())

# Distribusi PM2.5
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['PM2.5'], bins=50, kde=True)
plt.title('Distribusi Konsentrasi PM2.5')
plt.xlabel('PM2.5 (g/m)')
plt.ylabel('Frekuensi')
plt.show()

# Pola temporal
monthly_pm = combined_df.resample('M', on='date')['PM2.5'].mean()
plt.figure(figsize=(12, 6))
monthly_pm.plot()
plt.title('Tren Bulanan PM2.5 (2013-2017)')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata PM2.5')
plt.grid(True)
plt.show()

# Perbandingan antar stasiun
plt.figure(figsize=(12, 6))
sns.boxplot(x='station', y='PM2.5', data=combined_df)
plt.title('Distribusi PM2.5 per Stasiun')
plt.xticks(rotation=45)
plt.show()
```

# Analisis Tren PM2.5  

## 1. Pola Musiman  
Tingkat PM2.5 menunjukkan lonjakan musiman yang jelas, dengan konsentrasi lebih tinggi selama musim dingin. Hal ini kemungkinan disebabkan oleh:  
- Peningkatan aktivitas pemanasan  
- Kondisi udara yang stagnan  

## 2. Tren Stasiun  
Berbagai stasiun pemantauan menunjukkan pola yang serupa, yang mengindikasikan bahwa tingkat PM2.5 lebih dipengaruhi oleh faktor regional daripada sumber polusi yang terisolasi.  

## 3. Puncak Anomali  
Terdapat lonjakan PM2.5 yang tidak biasa pada periode tertentu, seperti:  
- Awal tahun 2014  
- Awal tahun 2016  

Kemungkinan penyebabnya meliputi:  
- Badai debu  
- Episode polusi tinggi akibat aktivitas industri atau kebakaran hutan  

## 4. Penurunan Tahunan  
Meskipun terdapat fluktuasi musiman, tren PM2.5 secara keseluruhan menunjukkan kemungkinan adanya perbaikan kualitas udara dari waktu ke waktu. Hal ini bisa disebabkan oleh:  
- Kebijakan regulasi yang lebih ketat  
- Pengurangan emisi dari sektor industri dan transportasi  

```python
# Heatmap korelasi
plt.figure(figsize=(10, 8))
corr_matrix = combined_df[['PM2.5', 'TEMP', 'PRES', 'WSPM', 'DEWP']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasi Parameter Udara')
plt.show()

# Analisis klaster stasiun
station_stats = combined_df.groupby('station').agg(
    avg_pm25=('PM2.5', 'mean'),
    max_pm25=('PM2.5', 'max'),
    pollution_days=('PM2.5', lambda x: (x > 150).sum())
).sort_values('avg_pm25', ascending=False)

station_stats['risk_level'] = pd.qcut(station_stats['avg_pm25'], 
                                     q=3, 
                                     labels=['Rendah', 'Sedang', 'Tinggi'])

print("Klasifikasi Risiko Polusi per Stasiun:")
station_stats
```

```python
1. **Tren PM2.5** menunjukkan pola musiman dengan puncak di musim dingin karena peningkatan penggunaan pemanas batubara dan kondisi inversi atmosfer.
2. **Faktor meteorologi** yang signifikan:
   - Suhu rendah berkorelasi dengan polusi tinggi.
   - Kecepatan angin rendah berhubungan dengan akumulasi polutan.
3. **Rekomendasi**:
   - Fokus kontrol polusi di stasiun dengan risiko tinggi.
   - Penguatan regulasi emisi selama musim dingin.
   - Sistem peringatan dini berbasis prediksi meteorologi.
```
